#### Finding Duplicated columns and removing them

```python
def getDuplicateColumns(df):
    '''
    Get a list of duplicate columns.
    It will iterate over all the columns in dataframe and find the columns whose contents are duplicate.
    :param df: Dataframe object
    :return: List of columns whose contents are duplicates.
    '''
    duplicateColumnNames = set()
    # Iterate over all the columns in dataframe
    for x in range(df.shape[1]):
        # Select column at xth index.
        col = df.iloc[:, x]
        # Iterate over all the columns in DataFrame from (x+1)th index till end
        for y in range(x + 1, df.shape[1]):
            # Select column at yth index.
            otherCol = df.iloc[:, y]
            # Check if two columns at x 7 y index are equal
            if col.equals(otherCol):
                duplicateColumnNames.add(df.columns.values[y])
 
    return list(duplicateColumnNames)
```

### Dimensionality reduction witg PCA

```python
from sklearn.decomposition import PCA
```

```python
pca = PCA()
principalComponents = pca.fit_transform(
    df.drop( "TARGET", axis=1)
)
variance = pca.explained_variance_ratio_  # calculate variance ratios
var = np.cumsum(np.round(pca.explained_variance_ratio_, decimals=3) * 100)
```

```python
test=test.drop('TARGET',axis=1)
```

```python
clf=ExtraTreesClassifier(random_state=7)
```

```python
selector=clf.fit(X_train,y_train)
```

#### Plotting feature importnace according to ExtraTreeClassifier

as ExtraTreeClassifier is not expensive to train we just used it for finding important features. Later on we will use those features to train the final model.

```python
feat_imp = pd.Series(clf.feature_importances_, index = X_train.columns.values).sort_values(ascending=False)
feat_imp[:40].plot(kind='bar', title='Feature Importances according to ExtraTreesClassifier', figsize=(15, 12))
plt.ylabel('Feature Importance Score')
plt.subplots_adjust(bottom=0.3)
plt.savefig('1.png')
plt.show()
```

As could be seen below we selected only 37 features out of 370 and now we can use decision trees like random forest or xgb 

```python
fs = SelectFromModel(selector, prefit=True)

X_train = fs.transform(X_train)
X_test = fs.transform(X_test)
test = fs.transform(test)
```
